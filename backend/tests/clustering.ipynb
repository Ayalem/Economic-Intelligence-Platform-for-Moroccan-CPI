```python
import pandas as pd
import numpy as np
```

```python
def preparing_df(df):
  def verify_date(col):
    try:
        pd.to_datetime(col,format="%Y-%m-%d %H:%M:%S")
        return True
    except:
        return False
  city_map={
    1:'Agadir',
    2:'Casablanca',
    3:'Fes',
    4:'Kenitra',
    5:'Marrackech',
    6:'Oujda',
    7:'Rabat',
    8:'Tetouan',
    9:'Meknes',
    10:'Tanger',
    11:'Laayoune',
    12:'Dakhla',
    13:'Guelmim',
    14:'Settat',
    15:'Safi',
    16:'Beni Mellal',
    17:'Al-Houcima',
    18:'Errachidia',
    99:"National"
}
  date_cols=[col for col in df.columns  if verify_date(col)]
  df=df.melt(id_vars=["VILLE","coicop","libelle"],var_name="Date",value_vars=date_cols,value_name="IPC")
  df["VILLE"]=df["VILLE"].map(city_map)
  df["Date"]=pd.to_datetime(df["Date"],errors="coerce").astype(str)
  return df
```

```python
df_now=preparing_df(df)
df_now_cat=df_now[df_now['coicop'].str.len()==2]
```

```python
df_now_cat.head()
```

```python
df_now_cat=df_now_cat[df_now_cat['VILLE']=='National']
df_now_cat=df_now_cat.drop(['VILLE','coicop'],axis=1)
df_now_cat.head()
```

```python
df_final=df_now_cat.pivot(index='libelle',columns='Date',values='IPC')
df_final.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
sns.scatterplot(data =df_now_cat, x ='Date', y = 'IPC', hue = 'libelle')
plt.xticks(rotation=45)
plt.show()
```

```python
X=df_final.values
X
```

```python
from sklearn.preprocessing import StandardScaler
#normalize data:
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
print(X_scaled.shape)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
# Fit model to points
model.fit(X_scaled)
labels = model.labels_
print("labels:",labels)

```

```python
df_final['cluster']=labels
df_final.head()
```

```python
df_final.reset_index(inplace=True)
```

```python
df_final.head(12)
```

```python
df_cluster=df_final.groupby('cluster')['libelle'].value_counts()
df_cluster
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
X_scaled
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca
```

```python
labels = model.labels_

# Plot points colored by cluster
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='Set2', alpha=0.6)

# Plot centroids (in PCA space)
centroids = pca.transform(model.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=50, c='black', label='Centroids')

plt.title("KMeans Clustering with PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.show()
```
